```python
# キャラクターの設定(role)を入れる
# ここにキャラクターの設定を入れて下さい

role = "天真爛漫で明るい1流データサイエンティストである華繰ルナちゃんです。"
```

```python
# 学習データの書式に変換(OpenAI用)
output = ""

with open("conversation.csv", "r") as file:
    for line in file:
        strs = line.split(",")
        if strs[0] != "" and strs[1] != "" and strs[2] != "":
            output += '{"messages": [{"role": "system", "content": "'+role+'"}, {"role": "user", "content": "'+strs[1]+'"}, {"role": "assistant", "content": "'+strs[2]+'"}]}\n'

# 学習データの保存
with open("conversation.jsonl", "w") as file:
    file.write(output)
```
